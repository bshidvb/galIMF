```python
from scipy.integrate import quad
import glob
import numpy as np
import matplotlib.pyplot as plt
import os
%matplotlib inline

def load_file_paths(base_dir):
    file_paths = []
    for root, dirs, files in os.walk(base_dir):
        for file in files:
            if file == "N_over_O_time.txt":
                file_paths.append(os.path.join(root, file))
    return file_paths

def load_data_with_names(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
    data_dict = {}
    for i in range(len(lines)):
        line = lines[i].strip()
        if line.startswith("#"):
            row_name = line[1:].strip()
            if i + 1 < len(lines):
                data = list(map(float, lines[i + 1].strip().split()))
                data_dict[row_name] = data
    return data_dict

def recalculate(N_over_O, O_over_H): # Recalculate logNO and logOH
    N_over_O = [i - 0.86 for i in N_over_O]
    O_over_H = [i + 8.696 for i in O_over_H]
    return N_over_O, O_over_H

def prepare_data(file_paths):
    for path in file_paths:
        data = load_data_with_names(path)
        log_time, N_over_O = data['log_time_axis'], data['gas_N_over_O_list']
        # timestep = data['time step list:']
    #     logNO_rec, logOH_rec = recalculate(logNO, logOH)
    #     smooth_logOH, smooth_logNO = fit(logOH_rec, logNO_rec)
    # return smooth_logOH, smooth_logNO
    return log_time, N_over_O
```
