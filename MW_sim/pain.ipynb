```python
import pandas as pd
mw_data = pd.read_csv("./MW_data/Spite2005_metal-poor-giants.txt", delimiter="|")
FeH_spite = mw_data['[Fe/H]']
NH_spite = mw_data['adopt. [N/H]']
OH_spite = mw_data['[O/H]']

# Calculate [N/O] = [N/H] - [O/H]
NO_spite = NH_spite - OH_spite

import matplotlib.pyplot as plt

plt.scatter(OH_spite, NO_spite)
plt.xlabel('[O/H]')
plt.ylabel('[N/O]')
plt.show()
```
